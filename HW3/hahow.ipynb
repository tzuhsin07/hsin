```python
import json 
import pandas as pd

data_list=[]
header = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36"
}
url = "https://api.hahow.in/api/products/search?category=COURSE&limit=24&page=0&sort=TRENDING"
for y in range(43):
    url = "https://api.hahow.in/api/products/search?category=COURSE&limit=24&page="
    url2 = "&sort=TRENDING"
    url3 = url+str(y)+url2
    
    res = re.get(url3,headers = header)

    res_json = res.json()

    for data1 in res_json['data']['courseData']['products']:
        data = {}
        title = data1['title']
        data["課程:"]=title
        price = data1["purchasePlan"]['price']
        data["價格:"]=price
        name = data1["owner"]['name']
        data["作者:"]=name
        num = data1["numSoldTickets"]
        data["參與人數:"]=num
        data_list.append(data)
        
df = pd.DataFrame(data_list)
df.to_excel('course09.data.xlsx', index=False)
```
