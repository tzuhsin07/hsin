```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import csv
url = "https://www.twse.com.tw/rwd/zh/afterTrading/MI_INDEX?date=20231027&type=0099P&response=html"
response = requests.get(url)
soup = BeautifulSoup(response.text,"html.parser")
stock = soup.find_all("tr",align="center",style="font-size:14px;")
#print(stock)

data_list = []  


for row in stock:
    cells = row.find_all('td')
    data = {  
        "證券代號": cells[0].text,
        "證券名稱": cells[1].text,
        "成交股數": cells[2].text,
        "成交金額": cells[3].text,
        "開盤價": cells[5].text,
        "最高價": cells[6].text,
        "最低價": cells[7].text,
        "收盤價": cells[8].text
    }
    data_list.append(data)


with open("stock.json2", "w", encoding="utf-8") as file:
    json.dump(data_list, file, ensure_ascii=False, indent=4)
    
df = pd.DataFrame(data_list)
df.to_excel('stock.data.xlsx', index=False)
```
