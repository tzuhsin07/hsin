```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 設定 ChromeDriver 路徑
chrome_driver_path = './chromedriver_win32/chromedriver.exe'

# 初始化 ChromeDriver 並設定 options
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Optional: Run in headless mode to avoid opening a visible browser window
options.add_argument('--disable-gpu')  # Avoid an error message when running in headless mode

# 使用 ChromeService 來設定 ChromeDriver 路徑
service = ChromeService(chrome_driver_path)

# 傳遞 service 到 ChromeDriver 的 service 參數
driver = webdriver.Chrome(service=service, options=options)

for i in range(1, 12):
    month = f"{i:02}"
    for x in range(1, 31):
        date = f"{x:02}"
        for y in range(1, 24):
            hour = f"{y:02}"
            for z in range(0, 60):
                mini = f"{z:02}"
                for s in range(0, 60):
                    sec = f"{s:02}"
                    for u in range(0, 99):
                        scale = f"{u:02}"
                        
                        url = f"https://scweb.cwa.gov.tw/zh-tw/earthquake/details/2023{month}{date}{hour}{mini}{sec}{scale}"
                        
                        # 使用 Selenium 開啟網頁
                        driver.get(url)
                        
                        # 使用 WebDriverWait 來等待特定元素出現
                        wait = WebDriverWait(driver, 10)
                        wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'eqReportHead')))
                        
                        # 取得網頁內容
                        page_source = driver.page_source
                        
                        # 直接打印網頁內容
                        print(page_source)

# 關閉瀏覽器
driver.quit()

```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 設定 ChromeDriver 路徑
chrome_driver_path = './chromedriver_win32/chromedriver.exe'

# 初始化 ChromeDriver 並設定 options
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Optional: Run in headless mode to avoid opening a visible browser window
options.add_argument('--disable-gpu')  # Avoid an error message when running in headless mode

# 使用 ChromeService 來設定 ChromeDriver 路徑
service = ChromeService(chrome_driver_path)

# 傳遞 service 到 ChromeDriver 的 service 參數
driver = webdriver.Chrome(service=service, options=options)

for i in range(1, 12):
    month = f"{i:02}"
    for x in range(1, 31):
        date = f"{x:02}"
        for y in range(1, 24):
            hour = f"{y:02}"
            for z in range(0, 60):
                mini = f"{z:02}"
                for s in range(0, 60):
                    sec = f"{s:02}"
                    for u in range(0, 99):
                        scale = f"{u:02}"
                        
                        url = f"https://scweb.cwa.gov.tw/zh-tw/earthquake/details/2023{month}{date}{hour}{mini}{sec}{scale}"
                        
                        # 使用 Selenium 開啟網頁
                        driver.get(url)
                        
                        # 使用 WebDriverWait 來等待特定元素出現
                        wait = WebDriverWait(driver, 20)  # 增加等待時間
                        wait.until(EC.visibility_of_element_located((By.CLASS_NAME, 'eqReportHead')))
                        
                        # 取得網頁內容
                        page_source = driver.page_source
                        
                        # 直接打印網頁內容
                        print(page_source)

# 關閉瀏覽器
driver.quit()
```

```python
from selenium import webdriver

# 設定 ChromeDriver 路徑
chrome_driver_path = './chromedriver_win32/chromedriver.exe'

# 設定 ChromeOptions
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')  # Optional: Run in headless mode to avoid opening a visible browser window
chrome_options.add_argument('--disable-gpu')  # Avoid an error message when running in headless mode

# 初始化 ChromeDriver 並設定 options 和 executable_path
driver = webdriver.Chrome(executable_path=chrome_driver_path, options=chrome_options)

for i in range(1, 12):
    month = f"{i:02}"
    for x in range(1, 31):
        date = f"{x:02}"
        for y in range(1, 24):
            hour = f"{y:02}"
            for z in range(0, 60):
                mini = f"{z:02}"
                for s in range(0, 60):
                    sec = f"{s:02}"
                    for u in range(0, 99):
                        scale = f"{u:02}"

                        url = f"https://scweb.cwa.gov.tw/zh-tw/earthquake/details/2023{month}{date}{hour}{mini}{sec}{scale}"

                        # 使用 Selenium 開啟網頁
                        driver.get(url)

                        # 取得網頁內容
                        page_source = driver.page_source

                        # 直接打印網頁內容
                        print(page_source)

# 關閉瀏覽器
driver.quit()

```
