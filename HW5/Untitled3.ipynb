```python
import requests
from bs4 import BeautifulSoup

header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36"
}

for i in range(1, 12):
    month = f"{i:02}"
    for x in range(1, 31):
        date = f"{x:02}"
        for y in range(1, 24):
            hour = f"{y:02}"
            for z in range(0, 60):
                mini = f"{z:02}"
                for s in range(0, 60):
                    sec = f"{s:02}"
                    for u in range(0, 99):
                        scale = f"{u:02}"
                        url = f"https://scweb.cwa.gov.tw/zh-tw/earthquake/details/2023{month}{date}{hour}{mini}{sec}{scale}"
                        if url == f"https://scweb.cwa.gov.tw/zh-tw/earthquake/details/2023{month}{date}{hour}{mini}{sec}{scale}":
                            response = requests.get(url,headers=header)
                            print(response.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")
articles = soup.find_all("div", class_="eqResultBoxBg clear")

coordinates = []

for article in articles:
    text_elements = article.find_all("div", class_="text")
    for element in text_elements:
        text = element.get_text()
        if "北緯" in text or "東經" in text:
            coordinates.append(text)

eq_magnitude = soup.find("span", text="芮氏規模：")
if eq_magnitude:
    magnitude = eq_magnitude.find_next("div", class_="text")
    if magnitude:
        print("規模:", magnitude.get_text())


for coordinate in coordinates:
    print(coordinate)
```

```python
import requests
from bs4 import BeautifulSoup

header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36"
}

current_year = 2023  # 可以动态获取当前年份

for month in range(1, 12):
    for day in range(1, 31):
        for hour in range(1, 24):
            for minute in range(0, 60):
                for second in range(0, 60):
                    for millisecond in range(0, 99):
                        scale = f"{millisecond:02}"
                        url = f"https://scweb.cwa.gov.tw/zh-tw/earthquake/details/{current_year:04}{month:02}{day:02}{hour:02}{minute:02}{second:02}{scale}"
                        try:
                            response = requests.get(url, headers=header)
                            response.raise_for_status()  # 检查是否有错误的响应
                            print(response.text)
                        except requests.exceptions.RequestException as e:
                            print(f"访问 {url} 时出错: {e}")
```
