```python
import requests
from bs4 import BeautifulSoup

header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36"
}

url = "https://scweb.cwa.gov.tw/zh-tw/earthquake/details/2023012923151832"
response = requests.get(url, verify=False, headers=header)
print(response.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text,"html.parser")
articles = soup.find_all("div",class_="eqResultBoxBg clear")

for article in articles:
    text_elements = article.find_all("div", class_="text")
    for element in text_elements:
        if "北緯" in element.get_text() or "東經" in element.get_text():
            print(element.get_text())
eq_magnitude = soup.find("span", text="芮氏規模：")
if eq_magnitude:
    magnitude = eq_magnitude.find_next("div", class_="text")
    if magnitude:
        print(magnitude.get_text())
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")
articles = soup.find_all("div", class_="eqResultBoxBg clear")

coordinates = []

for article in articles:
    text_elements = article.find_all("div", class_="text")
    for element in text_elements:
        text = element.get_text()
        if "北緯" in text or "東經" in text:
            coordinates.append(text)

print("經緯度信息：")
for coordinate in coordinates:
    print(coordinate)

eq_magnitude = soup.find("span", text="芮氏規模：")
if eq_magnitude:
    magnitude = eq_magnitude.find_next("div", class_="text")
    if magnitude:
        print("芮氏規模：" + magnitude.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36"
}

for i in range(7, 8):
    month = f"{i:02}"
    for x in range(1, 31):
        date = f"{x:02}"
        for y in range(1, 24):
            hour = f"{y:02}"
            for z in range(0, 60):
                mini = f"{z:02}"
                for s in range(0, 60):
                    sec = f"{s:02}"
                    for u in range(0, 99):
                        scale = f"{u:02}"
                        url = f"https://scweb.cwa.gov.tw/zh-tw/earthquake/details/2023{month}{date}{hour}{mini}{sec}{scale}"
                        response = requests.get(url, headers=header)
                        print(response.text)
```
