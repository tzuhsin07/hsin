```python
import pandas as pd
df_earthquake = pd.read_csv('earthquake.csv',encoding="big5")
df_earthquake.head()
```

```python
import folium
m = folium.Map([23.5, 121], zoom_start=7,tiles='OpenStreetMap')
folium.Marker(location=[df_earthquake['Unnamed: 3'][1], df_earthquake['Unnamed: 2'][1]]).add_to(m)
folium.Marker(location=[df_earthquake['Unnamed: 3'][2], df_earthquake['Unnamed: 2'][2]]).add_to(m)
folium.Marker(location=[df_earthquake['Unnamed: 3'][3], df_earthquake['Unnamed: 2'][3]]).add_to(m)

m

```

```python
m0 = folium.Map([23.5, 121], zoom_start=7,tiles='OpenStreetMap')
for index, row in df_earthquake.iterrows():
    folium.Marker(location=[row['Unnamed: 3'],row['Unnamed: 2']]).add_to(m0)
m0
```

```python
m1 = folium.Map([23.5, 121], zoom_start=7)
from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(m1)

for index, row in df_earthquake.iterrows():
    if row['Unnamed: 5'] >=4.5 :
        information = str(row['Unnamed: 1'])+'<br>'+str(row['Unnamed: 5'])+'<br>'+str(row['Unnamed: 6'])
        folium.Marker(
            location=[row['Unnamed: 3'],row['Unnamed: 2']],
            
            popup=information).add_to(marker_cluster)
        
m1
```
