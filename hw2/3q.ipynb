```python
#將資料化為易讀形式並印出
import json
import pprint as pp

with open('covid.json', encoding='utf-8') as file:
    data = json.load(file)

pp.pprint(data, sort_dicts=False)
```

# Q1 輸入日期，並顯示當日確診人數以及總確診人數

```python
Q1= input("請輸入日期(2022-01-01~2022-12-31)：")
found = False  # 用來標記是否找到了匹配的日期

for data2 in data:
    if Q1 in data2['個案研判日']:
        print(f"日期: {data2['個案研判日']} 確診人數: {data2['新增確診人數']} 總確診人數:{data2['total']}")
        found = True

if not found:
    print("找不到該日期資料")
```

# Q2 輸入一個日期區間，計算區間內確診人數在總確診人數的比率

```python
Q2 = input("輸入日期")
Q4 = input("輸入日期")

found = False  

for data3 in data:
    if Q2 in data3['個案研判日']:
        num1 = data3['新增確診人數']
        total1 = data3['total']
    if Q4 in data3['個案研判日']:
        num2 = data3['新增確診人數']
        total2 = data3['total']
        found = True
if not found:
    print("請輸入正確日期")
if num1>num2:
    Proportion=(num1-num2)
    print(f"{Q4}至{Q2}的新增確診人數為{Proportion}")
else:
    Proportion=(num2-num1)
    print(f"{Q2}至{Q4}的新增確診人數為{Proportion}")
```

# Q3輸入日期，顯示各年齡層的第一接種率(以長條圖顯示)

```python
import json
import pprint as pp

with open('city_vaccination.json', encoding='utf-8') as file:
    vaccination = json.load(file)

pp.pprint(vaccination, sort_dicts=False)
```

```python
Q3=input("輸入日期")

found = False
for data4 in vaccination:
    if Q3 in data4['發佈統計日期']:
        print(f"日期: {data4['發佈統計日期']} 第1劑接種率: {data4['第1劑接種率']}")
        found = True

if not found:
    print("請輸入正確日期")

    
```

```python
Q3 = input("輸入日期")
found_data = []  
for data4 in vaccination:
    if Q3 == data4['發佈統計日期']:
        found_data.append(data4)

if found_data:
    for index, data in enumerate(found_data, start=1):
        print(f"匹配数据 {index}:")
        print(f"日期: {data['發佈統計日期']} 第1劑接種率: {data['第1劑接種率']}")
        print("---")
        print(type(data['第1劑接種率']))
else:
    print("請輸入正確日期")
```

```python
import matplotlib.pyplot as plt
Q3 = input("輸入日期")
categories = ['ALL', '6month-4', '5-11', '12-17', '18-29', '30-49', '50-64', '65-74', '75']
values = [] 
found_data = []


for data4 in vaccination:
    if Q3 == data4['發佈統計日期']:
        found_data.append(data4)
if found_data:
    for i in range(len(categories)):
        value = found_data[i]['第1劑接種率']  
        values.append(value)
else:
    print("請輸入正確日期")
    


plt.bar(categories, values)
plt.yticks(range(0, 101, 10)) 

plt.title('vaccination rate')
plt.xlabel('age')
plt.ylabel('ratio')


plt.show()

```
